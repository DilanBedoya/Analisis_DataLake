```python
import requests
from bs4 import BeautifulSoup

def extraer_datos_fila(fila):
    datos = []
    for celda in fila.find_all('td'):
        # Encuentro todas las etiquetas <span> dentro de la fila etiqueta td y las elimino
        for span in celda.find_all('span'):
            span.decompose()
        # Extraigo el texto de la etiqueta td después de eliminar las etiquetas <span> y su contenido
        texto = celda.get_text(strip=True)
        # Remuevo los caracteres no deseados del html
        texto = texto.replace('\xa0', '').replace('\u2009', '').replace('&','')
        datos.append(texto)
    return datos

listadopeliculas = []

try:
    #extraigo los datos del url y los almaceno en soup
    url = 'https://es.wikipedia.org/wiki/Anexo:Pel%C3%ADculas_con_las_mayores_recaudaciones'
    pedido_obtenido = requests.get(url)
    html_obtenido = pedido_obtenido.text
    soup = BeautifulSoup(html_obtenido, 'html.parser')
    
    #obtengo los datos exclusivamente de la tabla con el classname wikitable
    tabla = soup.find('table', class_='wikitable')
    if tabla:
        #si encuentro la tabla entra al bucle y acciona la funcion extraer datos fila y manda la fila despues retorna los datos fila y los almacena
        for fila in tabla.find_all('tr'):
            datos_fila = extraer_datos_fila(fila)
            if datos_fila:
                listadopeliculas.append(datos_fila)

except Exception as e:
    print(f'Ocurrió un error: {e}')

# Imprimir la lista de películas obtenidas
for pelicula in listadopeliculas:
    print(pelicula)
```

```python
#Creamos el archivo .db para sqlite y establecemos conexion
import sqlite3
conexion = sqlite3.connect('databasepeliculasglobales.db')
cursor= conexion.cursor()

#Creamos la tabla para el primer dataset (webscraping)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS PELICULASDATA1 (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        Nombre_de_película TEXT,
        Recaudacion_mundial TEXT,
        Taquilla_EEUU TEXT,
        Taquilla_fuera_de_EEUU TEXT,
        Presupuesto TEXT,
        Distribuidora TEXT,
        Año_de_estreno TEXT,
        Director TEXT
    )
''')


#Creamos la tabla para el segundo dataset
cursor.execute('''
    CREATE TABLE IF NOT EXISTS PELICULASDATA2 (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        names TEXT,
        date_x DATE,
        Score TEXT,
        genre TEXT,
        overview TEXT,
        crew TEXT,
        orig_title TEXT,
        status TEXT,
        orig_lang TEXT,
        budget_x TEXT,
        revenue TEXT,
        country TEXT
    )
''')


#Creamos la tabla para el segundo dataset
cursor.execute('''
    CREATE TABLE IF NOT EXISTS PELICULASDATA3 (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        imdbID TEXT,
        title TEXT,
        year DATE,
        rating TEXT,
        genre TEXT,
        released DATE,
        director TEXT,
        writer TEXT,
        cast TEXT,
        metacritic TEXT,
        imdbRating TEXT,
        imdbVotes TEXT,
        poster TEXT,
        language TEXT,
        country TEXT,
        lastupdated TEXT,
        type TEXT
    )
''')



conexion.commit()

for peliculas in listadopeliculas:
    cursor.execute('''
        INSERT INTO PELICULASDATA1 (Nombre_de_película, Recaudacion_mundial, Taquilla_EEUU, Taquilla_fuera_de_EEUU, Presupuesto, Distribuidora, Año_de_estreno, Director) VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    ''', (peliculas[0], peliculas[1], peliculas[2], peliculas[3], peliculas[4], peliculas[5], peliculas[6], peliculas[7]))


# Insertar los datos del Dataset2 en la tabla de la base de datos2
lectura.to_sql('PELICULASDATA2', conexion, if_exists='append', index=False)


# Insertar los datos del Dataset3 en la tabla de la base de datos3
lectura2.to_sql('PELICULASDATA3', conexion, if_exists='append', index=False)

conexion.commit()

conexion.close()
```
