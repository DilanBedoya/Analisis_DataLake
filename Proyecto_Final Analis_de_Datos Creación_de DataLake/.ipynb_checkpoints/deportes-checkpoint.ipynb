```python
import pandas as pd
#hacer df del dataset de campeones olimpicos
df_1 = pd.read_csv("dataset_olympics.csv") #leer el dataset como dataframe
df_1
```

```python
df_1.isnull().sum() #Verificar nulos
```

```python
#Calcular el promedio de la columna Age

promedio_age = round(df_1['Age'].mean(),0)
promedio_age
```

```python
#Calcular el promedio de la columna Height    

promedio_Height = round(df_1['Height'].mean(),0)
promedio_Height
```

```python
#Calcular el promedio de la columna Weight    

promedio_Weight = round(df_1['Weight'].mean(),0)
promedio_Weight
```

```python
#Llenar datos nulos con promedios y valor sin detalles

llenar_age = {'Age': '26.0'}
llenar_Height = {'Height': '176.0'}
llenar_Weight = {'Weight': '71.0'}
llenar_Medal = {'Medal': 'without details'}

df_1.fillna(llenar_age, inplace=True) #Modificar el archivo sin crear una copia
df_1.fillna(llenar_Height, inplace=True) #Modificar el archivo sin crear una copia
df_1.fillna(llenar_Weight, inplace=True) #Modificar el archivo sin crear una copia
df_1.fillna(llenar_Medal, inplace=True) #Modificar el archivo sin crear una copia
```
