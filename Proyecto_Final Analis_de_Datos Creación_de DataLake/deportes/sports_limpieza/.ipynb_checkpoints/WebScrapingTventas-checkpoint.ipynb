```python
import requests
from bs4 import BeautifulSoup
import pymongo
#https://www.tventas.com/
#Primero obtengo el html
#URL = 'https://www.tventas.com/'


#Segundo parsear el HTML


listadoimagenes=[]
mas_paginas=True
pagina = 1
try:
    while(mas_paginas):
        pedido_obtenido= requests.get(f'https://www.tventas.com/nuevos-productos?p={pagina}')
        html_obtenido = pedido_obtenido.text
        soup = BeautifulSoup(html_obtenido,'html.parser')
        #enlace_paginas = requests.get(f'https://www.tventas.com/nuevos-productos?p={pagina}').text
        for seccion in soup.find_all('img', title=True):
            imagenes={
                'title':seccion.get('title'),
                'src':seccion.get('src'),
                'alt':seccion.get('alt')
            }
            listadoimagenes.append(imagenes)
        print(f'HTML de la pagina{pagina}:\n[html_obtenido]')
        siguiente_pagina = soup.select_one('.icon-chevron-right')
        print(f'paginas exploradas {pagina}')
        if(pagina<=10):
            pagina += 1
        else:
            mas_paginas = False
except Exception as e:
    print(f'Ocurrio un error:{e}')


import sqlite3
conexion = sqlite3.connect('prueba2.db')
cursor= conexion.cursor()


cursor.execute('''
    CREATE TABLE IF NOT EXISTS IMAGENES (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        title TEXT,
        src TEXT,
        alt TEXT
    )
''')
conexion.commit()

for imagenes in listadoimagenes:
    cursor.execute('''
        INSERT INTO IMAGENES (title, src, alt) VALUES (?, ?, ?)
    ''', (imagenes['title'], imagenes['src'], imagenes['alt']))

conexion.commit()


cursor.execute("SELECT* FROM IMAGENES")
resultados= cursor.fetchall()
print(resultados)


conexion.close()
```

```python
import requests
from bs4 import BeautifulSoup
import pyodbc

# Obtener las imágenes desde el sitio web
listadoimagenes=[]
mas_paginas=True
pagina = 1
try:
    while(mas_paginas):
        pedido_obtenido= requests.get(f'https://www.tventas.com/nuevos-productos?p={pagina}')
        html_obtenido = pedido_obtenido.text
        soup = BeautifulSoup(html_obtenido,'html.parser')
        #enlace_paginas = requests.get(f'https://www.tventas.com/nuevos-productos?p={pagina}').text
        for seccion in soup.find_all('img', title=True):
            imagenes={
                'title':seccion.get('title'),
                'src':seccion.get('src'),
                'alt':seccion.get('alt')
            }
            listadoimagenes.append(imagenes)
        print(f'HTML de la pagina{pagina}:\n[html_obtenido]')
        siguiente_pagina = soup.select_one('.icon-chevron-right')
        print(f'paginas exploradas {pagina}')
        if(pagina<=10):
            pagina += 1
        else:
            mas_paginas = False
except Exception as e:
    print(f'Ocurrio un error:{e}')


# Conectar a la base de datos SQL Server
conexion_sql_server = pyodbc.connect(
    'DRIVER={SQL Server};'
    'SERVER=ALEXIS\SQLEXPRESS;'
    'DATABASE=Tventas;'
    'Trusted_Connection=yes;'
)

cursor_sql_server = conexion_sql_server.cursor()

# Crear la tabla si no existe (corregido para SQL Server)
cursor_sql_server.execute('''
    IF NOT EXISTS (SELECT * FROM INFORMATION_SCHEMA.TABLES WHERE TABLE_NAME = 'IMAGENES')
    BEGIN
        CREATE TABLE IMAGENES (
            id INT PRIMARY KEY IDENTITY(1,1),
            title NVARCHAR(255),
            src NVARCHAR(255),
            alt NVARCHAR(255)
        )
    END
''')
conexion_sql_server.commit()

# Insertar los datos en la tabla IMAGENES
for imagenes in listadoimagenes:
    cursor_sql_server.execute('''
        INSERT INTO IMAGENES (title, src, alt) VALUES (?, ?, ?)
    ''', (imagenes['title'], imagenes['src'], imagenes['alt']))

conexion_sql_server.commit()

# Realizar una consulta y mostrar los resultados (corregido para SQL Server)
cursor_sql_server.execute("SELECT * FROM IMAGENES")
resultados_sql_server = cursor_sql_server.fetchall()
for row in resultados_sql_server:
    print(row)

# Cerrar la conexión
conexion_sql_server.close()
```
